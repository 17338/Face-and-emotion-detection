```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
test_image = cv2.imread('astro.jpg')

#Converting to grayscale
test_image_gray = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)

# Displaying the grayscale image
plt.imshow(test_image_gray, cmap='gray')
```

```python
haar_cascade_face = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
faces_rect = haar_cascade_face.detectMultiScale(test_image_gray, scaleFactor = 1.1, minNeighbors = 5);

print((faces_rect))
print('Faces found: ', len(faces_rect))
```

```python
def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
 for (x, y, w, h) in faces_rect:
        cv2.rectangle(test_image, (x, y), (x+w, y+h), (0, 255, 0), 1)

plt.imshow(convertToRGB(test_image))
```
